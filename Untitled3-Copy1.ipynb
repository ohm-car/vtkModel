```python
def gen_points():
    
    pt = list()
    
    for i in range(3,8):
        for j in range(3,8):
            for k in range(3,8):
#                 pt.append([i/10, j/10, k/10])
                pt.append([i, j, k])                
    return pt
```

```python
def getXYsFromMask(mask):
    
    xys = np.argwhere(mask == True)
    return xys, xys[:,0], xys[:,1]
```

```python
def genPointsFromNPArray(A, sequence):
    
    z = 0
    points = list()
    for i in sequence:
        
        frame = A[i]
        
        xys = getXYsFromMask(frame)[0]
        
        for xy in xys:
            pt = xy.tolist()
            pt.append(z)
            points.append(pt)
        
        z += 1
    return points
```

```python
def main():
    colors = vtkNamedColors()

    # create a rendering window and renderer
    ren = vtkRenderer()
    renWin = vtkRenderWindow()
    renWin.AddRenderer(ren)
    renWin.SetWindowName('PointSource')

    # create a renderwindowinteractor
    iren = vtkRenderWindowInteractor()
    iren.SetRenderWindow(renWin)

    # Create a point cloud
    points = vtk.vtkPoints()
    for point in genPointsFromNPArray(mask,seq):
        points.InsertNextPoint(point[0], point[1], point[2])
        
    polydata = vtk.vtkPolyData()
    polydata.SetPoints(points)
    
    src = vtkPointSource()
#     src.SetCenter(0, 0, 0)
#     src.SetNumberOfPoints(50)
#     src.SetRadius(5)
#     src.Update()
    src.SetOutput(polydata)

    # mapper
    mapper = vtkPolyDataMapper()
    mapper.setInputData(polydata)
#     mapper.SetInputConnection(src.GetOutputPort())
#     mapper.SetInputConnection(points.GetOutputPort())

    # actor
    actor = vtkActor()
    actor.SetMapper(mapper)
    actor.GetProperty().SetColor(colors.GetColor3d('Tomato'))
    actor.GetProperty().SetPointSize(4)

    # assign actor to the renderer
    ren.AddActor(actor)
    ren.SetBackground(colors.GetColor3d('DarkGreen'))

    # enable user interface interactor
    iren.Initialize()
    renWin.Render()
    iren.Start()


if __name__ == '__main__':
    main()
```
